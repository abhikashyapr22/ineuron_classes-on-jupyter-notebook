```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

```python
train = pd.read_csv(r'C:\Users\Abhishek Kashyap\Desktop\Swiss Re’s_Participants_dataset\test.csv')
```

```python
train.head()
```

```python
state_dummies = pd.get_dummies(train['state'])
train.drop('country',axis=1,inplace=True)
Carriageway_dummies = pd.get_dummies(train['Carriageway_Hazards'])
special_dummies = pd.get_dummies(train['Special_Conditions_at_Site'])
road_surface = pd.get_dummies(train['Road_Surface_Conditions'])
weather = pd.get_dummies(train['Weather_Conditions'])
light_dummies = pd.get_dummies(train['Light_Conditions'])
train['Date'] = pd.to_datetime(train['Date'])
train['month'] = train['Date'].dt.month
train['Time'] = pd.to_datetime(train['Time'])
train['Hour'] = train['Time'].dt.hour
train['minute'] = train['Time'].dt.minute
train.drop(['Date','Time'], axis=1, inplace=True)
road_type = pd.get_dummies(train['Road_Type'])
human_control = pd.get_dummies(train['Pedestrian_Crossing-Human_Control'])
physical_fac = pd.get_dummies(train['Pedestrian_Crossing-Physical_Facilities'])
```

```python
train.drop(['Road_Type','Light_Conditions',
       'Weather_Conditions', 'Road_Surface_Conditions',
       'Special_Conditions_at_Site', 'Carriageway_Hazards','Pedestrian_Crossing-Human_Control',
       'Pedestrian_Crossing-Physical_Facilities'], axis=1, inplace=True)
train.drop('state',axis=1, inplace=True)
```

```python
final = pd.concat([road_type,state_dummies,Carriageway_dummies,special_dummies,
                   road_surface,weather,light_dummies,human_control,physical_fac],axis=1)
df = pd.concat([train,final],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for col in ['Local_Authority_(Highway)','Did_Police_Officer_Attend_Scene_of_Accident']:
    df[col] = le.fit_transform(df[col])
```

```python
data = df.groupby(df['postcode']).agg('mean').reset_index()
```

```python
data.drop('Accident_ID', axis=1, inplace=True)
```

```python
data['postcode'] = le.fit_transform(data['postcode'])
data.head()
```

```python
data.to_csv('sub_file.csv')
```
